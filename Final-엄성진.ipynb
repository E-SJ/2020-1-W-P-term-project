```python
from bs4 import BeautifulSoup
import requests
from time import sleep
import csv

def init_date():
    fileMatrix=[]
    with open('date.csv', 'r') as fileRead:
        Reader=csv.reader(fileRead)
        for line in Reader: 
            fileMatrix.append(line)
        global date_list

        date_list=[]
        for i in range(len(fileMatrix)):
            date_list.append(fileMatrix[i][0])
init_date()  

#외국 확진자 / 실행시간 약 1분
def foreign_decide(): #(3/16~5/31)
    with open('foreign_decide.csv', 'w') as fileWrite:
        myWriter = csv.writer(fileWrite)
        myWriter.writerow(["날짜", "미국 확진자","중국 확진자","일본 확진자"])
        for k in range(16+30+31): 
            i=date_list.index("20200316")+k
            day=date_list[i]
            while 1:
                url='http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19NatInfStateJson?serviceKey=DOkrIeTx1pocIyIaRRjFmEJfYhCoJ7fS8Qx%2Bl7FuMmrhWH%2Fn6XgltOCsBZFuSmc7NT%2Fmj89%2FaE%2BFXDkkaX%2Bhtg%3D%3D&pageNo=1&numOfRows=10&startCreateDt='+day+'&endCreateDt='+day+'&'
                req=requests.get(url)
                html=req.text
                soup=BeautifulSoup(html, 'html.parser')
                result_num = soup.find_all('resultcode')
                if result_num[0].text=="00":
                    break
                sleep(0.5)

            nation_name = soup.find_all('nationnm') #국가명
            total_num = soup.find_all('totalcount') #결과 수
            decide_num = soup.find_all('natdefcnt') #누적 확진자 수

            usa_decide_list={}
            chn_decide_list={}
            jpn_decide_list={}
            
            #같은 날짜에 데이터가 달라 같은 날짜면 최신것으로 업데이트 하기 위하여 거꾸로 진행
            for i in range(int(total_num[0].text)-1,-1,-1): 
                if nation_name[i].text=="미국":
                    usa_decide_list[day]=decide_num[i].text
                elif nation_name[i].text=="중국":
                    chn_decide_list[day]=decide_num[i].text
                elif nation_name[i].text=="일본":
                    jpn_decide_list[day]=decide_num[i].text

            myWriter.writerow([day, usa_decide_list[day],chn_decide_list[day],jpn_decide_list[day]])

#외국 사망자 / 실행시간 약 1분
def foreign_death(): #(3/16~5/31)
    with open('foreign_death.csv', 'w') as fileWrite:
        myWriter = csv.writer(fileWrite)
        myWriter.writerow(["날짜", "미국 사망자","중국 사망자","일본 사망자"])
        for k in range(16+30+31): 
            i=date_list.index("20200316")+k
            day=date_list[i]
            while 1:
                url='http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19NatInfStateJson?serviceKey=DOkrIeTx1pocIyIaRRjFmEJfYhCoJ7fS8Qx%2Bl7FuMmrhWH%2Fn6XgltOCsBZFuSmc7NT%2Fmj89%2FaE%2BFXDkkaX%2Bhtg%3D%3D&pageNo=1&numOfRows=10&startCreateDt='+day+'&endCreateDt='+day+'&'
                req=requests.get(url)
                html=req.text
                soup=BeautifulSoup(html, 'html.parser')
                result_num = soup.find_all('resultcode')
                if result_num[0].text=="00":
                    break
                sleep(0.5)

            nation_name = soup.find_all('nationnm') #국가명
            total_num = soup.find_all('totalcount') #결과 수
            death_num=soup.find_all('natdeathcnt') #누적 사망자 수

            usa_death_list={}
            chn_death_list={}
            jpn_death_list={}
            
            #같은 날짜에 데이터가 달라 같은 날짜면 최신것으로 업데이트 하기 위하여 거꾸로 진행
            for i in range(int(total_num[0].text)-1,-1,-1): 
                if nation_name[i].text=="미국":
                    usa_death_list[day]=death_num[i].text
                elif nation_name[i].text=="중국":
                    chn_death_list[day]=death_num[i].text
                elif nation_name[i].text=="일본":
                    jpn_death_list[day]=death_num[i].text

            myWriter.writerow([day, usa_death_list[day],chn_death_list[day],jpn_death_list[day]])
            
#지하철 / 실행시간 약 1분
def metro(): #(1/1~5/31)
    with open('metro.csv', 'w') as fileWrite:
        myWriter = csv.writer(fileWrite)
        myWriter.writerow(["날짜", "지하철 이용량(승차)"])
        for i in range(len(date_list)): 
            day=date_list[i]
            while 1:
                url='http://openapi.seoul.go.kr:8088/48746d75413734393634666a6d464f/xml/CardSubwayStatsNew/1/1000/'+day
                req=requests.get(url)
                html=req.text
                soup=BeautifulSoup(html, 'html.parser')
                result_num = soup.find_all('code')
                if result_num[0].text=="INFO-000":
                    break
                sleep(0.3)
            temp_sum=0
            metro_item=soup.find_all("ride_pasgr_num")
            for k in range(len(metro_item)):
                temp_sum+=int(metro_item[k].text)
            myWriter.writerow([day, temp_sum])
            
#버스 / 실행시간 약 1시간 20분
def bus(): #(1/1~5/31)
    with open('bus.csv', 'w') as fileWrite:
        myWriter = csv.writer(fileWrite)
        myWriter.writerow(["날짜", "버스 이용량(승차)"])
        for i in range(len(date_list)): 
            day=date_list[i]
            first=1
            last=1000
            temp_sum=0
            while 1:
                while 1:
                    url='http://openapi.seoul.go.kr:8088/48746d75413734393634666a6d464f/xml/CardBusStatisticsServiceNew/'+str(first)+'/'+str(last)+'/'+day
                    req=requests.get(url)
                    html=req.text
                    soup=BeautifulSoup(html, 'html.parser')
                    result_num = soup.find_all('code')
                    if result_num[0].text=="INFO-000":
                        break
                    sleep(0.1)
                listcnt_num= soup.find_all('list_total_count')
                bus_item=soup.find_all("ride_pasgr_num")
                for k in range(len(bus_item)):
                    temp_sum+=int(bus_item[k].text)
                if last<int(listcnt_num[0].text) :
                    first+=1000
                    last+=1000
                else:
                    myWriter.writerow([day, temp_sum])
                    break      
```
